```python
#1 analysis
import matplotlib.pyplot as plt
import numpy as np

# Data from your findings
loyalty_categories = ['Regular', 'Silver', 'Gold']
average_satisfaction = [5.0082, 5.01145, 5.01303]
average_purchase_amount = [9639.98, 9628.78, 9621.49]

# Set the positions of the bars on the x-axis
x = np.arange(len(loyalty_categories))

# Set the width of the bars
bar_width = 0.35

fig, ax1 = plt.subplots(figsize=(10, 6))

# Create bars for satisfaction scores
bars1 = ax1.bar(x - bar_width/2, average_satisfaction, bar_width, label='Average Satisfaction Score', color='lightblue')

# Manually set the y-axis range for satisfaction scores to make the differences more pronounced
ax1.set_ylim(5.005, 5.015)

# Create another y-axis for the purchase amounts
ax2 = ax1.twinx()
bars2 = ax2.bar(x + bar_width/2, average_purchase_amount, bar_width, label='Average Purchase Amount', color='orange')

# Manually set the y-axis range for purchase amounts to make the differences more noticeable
ax2.set_ylim(9500, 9700)

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax1.set_xlabel('Loyalty Status')
ax1.set_ylabel('Average Satisfaction Score', color='blue')
ax2.set_ylabel('Average Purchase Amount ($)', color='orange')
ax1.set_title('Comparison of Average Satisfaction Scores and Purchase Amounts by Loyalty Status')
ax1.set_xticks(x)
ax1.set_xticklabels(loyalty_categories)
ax1.tick_params(axis='y', labelcolor='blue')
ax2.tick_params(axis='y', labelcolor='orange')

# Add a legend
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

# Function to label the bars with the exact values
def add_value_labels(bars, ax, axis_name):
    for bar in bars:
        height = bar.get_height()
        if axis_name == 'ax2':
            ax.annotate(f'{height:.2f}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom', color='black', fontsize=8)
        else:
            ax.annotate(f'{height:.5f}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom', color='black', fontsize=8)

# Call functions to label bars
add_value_labels(bars1, ax1, 'ax1')
add_value_labels(bars2, ax2, 'ax2')

# Set the layout to be more accommodating of the plot elements
fig.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show the plot
plt.show()
```

```python
#3rd Analysis
import matplotlib.pyplot as plt

# Data from your findings
regions = ['East', 'West', 'North', 'South']
average_purchase_amounts = [9615.41, 9619.21, 9673.36, 9648.79]  # Repeated values for simplification
loyalty_counts = {
    'East': {'Silver': 8974, 'Regular': 18102, 'Gold': 2998},
    'West': {'Gold': 2933, 'Silver': 8975, 'Regular': 18027},
    'North': {'Regular': 11932, 'Gold': 1994, 'Silver': 5992},
    'South': {'Silver': 6023, 'Gold': 1973, 'Regular': 12077},
}

# Nested pie chart for loyalty status distribution
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
axs = axs.flatten()  # Flatten to 1D array
for i, region in enumerate(regions):
    sizes = list(loyalty_counts[region].values())
    labels = loyalty_counts[region].keys()
    axs[i].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    axs[i].set_title(f'Loyalty Status in {region}')
plt.suptitle('Loyalty Status Distribution by Region')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
#4th Analysis

import matplotlib.pyplot as plt

# Data from your findings
age_groups = ['Below 22', '22 to 35', 'Above 35']
product_categories = ['Health', 'Food']
purchase_amounts = {
    'Health': [9742.61, 9636.99, 9367.76],
    'Food': [10118.1, 9639.34, 9791.7]
}

# Define age groups as x-axis and purchase amounts as y-axis data
x = range(len(age_groups))

# Plot lines with markers for each product category
for category, amounts in purchase_amounts.items():
    plt.plot(x, amounts, marker='o', label=category)

# Add labels, title, and custom x-axis tick labels, etc.
plt.xlabel('Age Group')
plt.ylabel('Average Purchase Amount ($)')
plt.title('Average Purchase Amount by Age Group and Product Category')
plt.xticks(x, age_groups)
plt.legend()

# Annotate data points on the plot
for category, amounts in purchase_amounts.items():
    for i, amount in enumerate(amounts):
        plt.text(x[i], amount, f'${amount}', ha='center', va='bottom')

plt.show()
```

```python
#2nd Analysis
import matplotlib.pyplot as plt

# Data from your findings
purchase_frequencies = ['Rare', 'Occasional', 'Frequent']
purchase_amounts = [481479182, 288748068, 193251834]
colors = ['gold', 'lightcoral', 'lightskyblue']  # Colors for each section

# Create a larger pie chart
fig, ax = plt.subplots(figsize=(10, 10))  # Increased figure size
ax.pie(purchase_amounts, labels=purchase_frequencies, colors=colors, autopct='%1.1f%%', startangle=90)

# Draw a circle at the center of the pie to make it look like a donut
centre_circle = plt.Circle((0, 0), 0.70, color='white')
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle and looks fine
ax.axis('equal')
ax.set_title('Total Purchase Amount by Buying Frequency')

# Show the plot
plt.show()
```
